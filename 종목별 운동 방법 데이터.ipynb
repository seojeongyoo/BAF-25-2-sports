#### 4. 운동 영상 링크

```python
import pandas as pd
import numpy as np
from urllib.parse import quote

# 1. 기존에 생성했던 데이터프레임이 있다고 가정


# 2. 유튜브 검색 링크를 생성하는 함수 정의
def generate_youtube_search_link(sport_name):
    """
    운동 종목 이름을 받아, '[운동이름] 기초'에 대한
    유튜브 검색 결과 페이지 URL을 생성합니다.
    """
    search_query = f"{sport_name} 기초"
    encoded_query = quote(search_query)
    base_url = "https://www.youtube.com/results?search_query="
    return base_url + encoded_query

# 3. 'sports_nm' 컬럼에 함수를 적용하여 'media_url' 컬럼 생성 (이 부분이 수정됨)
df2['media_url'] = df2['sports_nm'].apply(generate_youtube_search_link)

# 4. 결과 확인 (이 부분이 수정됨)
df2
```

### 데이터베이스 업로드

```python
# 4. Supabase 테이블 데이터 전체 교체 (삭제 후 삽입)
TABLE_NAME = 'exercise_methods'

# ⬇️ [수정] 이 부분의 주석(#)을 제거해야 합니다.
# df2가 이 코드 이전에 정의되어 있어야 합니다.
records = df2.to_dict(orient='records') 

try:
    # 1단계: 테이블의 모든 데이터 삭제
    print(f"🔄 '{TABLE_NAME}' 테이블의 기존 데이터 전체 삭제를 시도합니다...")
    
    delete_response = supabase.table(TABLE_NAME).delete().gte('exer_cd', 0).execute()
    
    if delete_response.data or (not delete_response.data and (hasattr(delete_response, 'error') and delete_response.error is None)):
        print(f"✅ 기존 데이터 {len(delete_response.data)}건 삭제 완료.")
    elif hasattr(delete_response, 'error') and delete_response.error:
         print(f"⚠️ 기존 데이터 삭제 중 경고 또는 오류: {delete_response.error}")
         raise Exception(f"Delete failed: {delete_response.error}")
    else:
         print(f"✅ 기존 데이터 0건 삭제 완료 (테이블이 이미 비어있음).")

    # 2단계: 새 데이터(records) 전체 삽입
    # (여기서 'records' 변수가 필요합니다)
    print(f"🔄 새 데이터 {len(records)}건 삽입을 시도합니다...")
    insert_response = supabase.table(TABLE_NAME).insert(records).execute()
    
    if insert_response.data:
        print(f"\n✅✅✅ {len(insert_response.data)}개의 새 데이터가 '{TABLE_NAME}' 테이블에 성공적으로 업로드되었습니다.")
    else:
        print(f"❌ 새 데이터 업로드 실패. 응답을 확인하세요: {insert_response}")

except Exception as e:
    print(f"\n❌ 전체 작업 중 오류 발생: {e}")
```
