```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import json

# ====================================================================
# 1. ì„¤ì • Ø¨Ø®Ø´ (Configuration Section) - ì´ì „ê³¼ ë™ì¼
# ====================================================================
SERVICE_KEY = "d88b1ae9ffdae01dc67a2b4073ec4a035567a3a2a436e495b271eaa02b4cc69f"
TARGET_LOCATION_CITY = "ì„œìš¸íŠ¹ë³„ì‹œ"
TARGET_LOCATION_DISTRICT = "ì†¡íŒŒêµ¬"
EXCEL_FILE_PATH = "ê¸°ìƒì²­41_ë‹¨ê¸°ì˜ˆë³´ ì¡°íšŒì„œë¹„ìŠ¤_ì˜¤í”ˆAPIí™œìš©ê°€ì´ë“œ_ê²©ì_ìœ„ê²½ë„(2411).xlsx"

# ====================================================================
# 2. í•¨ìˆ˜ Ø¨Ø®Ø´ (Functions Section)
# ====================================================================

def get_grid_coords(city, district, file_path):
    """Excel íŒŒì¼ì—ì„œ ì§€ì •ëœ ì§€ì—­ì˜ ê²©ì ì¢Œí‘œ (nx, ny)ë¥¼ ì°¾ìŠµë‹ˆë‹¤."""
    try:
        df = pd.read_excel(file_path)
        df.columns = df.columns.str.strip()
        location = df[(df['1ë‹¨ê³„'] == city) & (df['2ë‹¨ê³„'] == district)]
        if not location.empty:
            return location.iloc[0]['ê²©ì X'], location.iloc[0]['ê²©ì Y']
        else:
            print(f"ì˜¤ë¥˜: '{city} {district}' ìœ„ì¹˜ë¥¼ ì—‘ì…€ íŒŒì¼ì—ì„œ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return None, None
    except FileNotFoundError:
        print(f"ì˜¤ë¥˜: ì—‘ì…€ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê²½ë¡œì™€ íŒŒì¼ ì´ë¦„ì„ í™•ì¸í•˜ì„¸ìš”: {file_path}")
        return None, None
    except Exception as e:
        print(f"íŒŒì¼ì„ ì½ëŠ” ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
        return None, None

# ======[ì´ í•¨ìˆ˜ê°€ ìˆ˜ì •ë˜ì—ˆìŠµë‹ˆë‹¤]======
def get_latest_base_time():
    """
    NaN ë¬¸ì œë¥¼ í•´ê²°í•˜ê¸° ìœ„í•´, í•­ìƒ ë‹¹ì¼ ì „ì²´ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ìˆë„ë¡
    base_timeì„ '0200'(ìƒˆë²½ 2ì‹œ ë°œí‘œ)ìœ¼ë¡œ ê³ ì •í•©ë‹ˆë‹¤.
    """
    now = datetime.now()
    base_date = now.strftime('%Y%m%d')
    base_time = "0200"  # í•˜ë£¨ ì „ì²´ ë°ì´í„°ë¥¼ ë°›ê¸° ìœ„í•´ ê°€ì¥ ì´ë¥¸ ì˜ˆë³´ ì‹œê°„ìœ¼ë¡œ ê³ ì •

    print(f"ì „ì²´ ë°ì´í„° ì¡°íšŒë¥¼ ìœ„í•´ ì˜ˆë³´ ë°œí‘œ ì‹œê°ì„ '{base_date} {base_time}'ìœ¼ë¡œ ê³ ì •í•©ë‹ˆë‹¤.")
    return base_date, base_time


def get_weather_forecast(service_key, nx, ny):
    endpoint = "http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getVilageFcst"
    base_date, base_time = get_latest_base_time()

    import urllib.parse
    service_key_encoded = urllib.parse.quote_plus(service_key)

    params = {
        'serviceKey': service_key_encoded,
        'pageNo': '1',
        'numOfRows': '1000',
        'dataType': 'JSON',
        'base_date': base_date,
        'base_time': base_time,
        'nx': str(nx),
        'ny': str(ny)
    }

    try:
        response = requests.get(endpoint, params=params)
        response.raise_for_status()
        data = response.json()
        result_code = data.get('response', {}).get('header', {}).get('resultCode')
        if result_code == '00':
            return data['response']['body']['items']['item']
        else:
            error_msg = data.get('response', {}).get('header', {}).get('resultMsg')
            print(f"API ì˜¤ë¥˜: {error_msg} (ì½”ë“œ: {result_code})")
            return None
    except requests.exceptions.RequestException as e:
        print(f"HTTP ìš”ì²­ ì˜¤ë¥˜: {e}")
        return None


def create_forecast_dataframe(forecast_data):
    """
    API ì‘ë‹µ ë°ì´í„°ë¥¼ pandas DataFrameìœ¼ë¡œ ë³€í™˜í•˜ê³  NaN ê°’ì„ ì±„ì›ë‹ˆë‹¤.
    """
    if not forecast_data:
        print("ë‚ ì”¨ ì •ë³´ê°€ ì—†ì–´ ë°ì´í„°í”„ë ˆì„ì„ ìƒì„±í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return pd.DataFrame()

    category_map = {
        'POP': 'ê°•ìˆ˜í™•ë¥ ', 'PTY': 'ê°•ìˆ˜í˜•íƒœ', 'PCP': '1ì‹œê°„ ê°•ìˆ˜ëŸ‰', 'REH': 'ìŠµë„',
        'SNO': '1ì‹œê°„ ì‹ ì ì„¤', 'SKY': 'í•˜ëŠ˜ìƒíƒœ', 'TMP': '1ì‹œê°„ ê¸°ì˜¨',
        'TMN': 'ì¼ ìµœì €ê¸°ì˜¨', 'TMX': 'ì¼ ìµœê³ ê¸°ì˜¨', 'WSD': 'í’ì†'
    }
    sky_map = {'1': 'ë§‘ìŒ â˜€ï¸', '3': 'êµ¬ë¦„ë§ìŒ â˜ï¸', '4': 'íë¦¼ ğŸŒ¥ï¸'}
    pty_map = {'0': 'ì—†ìŒ', '1': 'ë¹„ â˜”', '2': 'ë¹„/ëˆˆ', '3': 'ëˆˆ â˜ƒï¸', '4': 'ì†Œë‚˜ê¸° ğŸŒ¦ï¸'}

    df = pd.DataFrame(forecast_data)
    df['category'] = df['category'].map(category_map)
    
    is_sky = df['category'] == 'í•˜ëŠ˜ìƒíƒœ'
    df.loc[is_sky, 'fcstValue'] = df.loc[is_sky, 'fcstValue'].map(sky_map)
    
    is_pty = df['category'] == 'ê°•ìˆ˜í˜•íƒœ'
    df.loc[is_pty, 'fcstValue'] = df.loc[is_pty, 'fcstValue'].map(pty_map)
    
    df_pivot = df.pivot_table(
        index=['fcstDate', 'fcstTime'], 
        columns='category', 
        values='fcstValue',
        aggfunc='first'
    ).reset_index()
    
    df_pivot.rename(columns={'fcstDate': 'ì˜ˆë³´ì¼ì', 'fcstTime': 'ì˜ˆë³´ì‹œê°'}, inplace=True)
    
    df_pivot['ì¼ ìµœì €ê¸°ì˜¨'] = df_pivot.groupby('ì˜ˆë³´ì¼ì')['ì¼ ìµœì €ê¸°ì˜¨'].transform(lambda x: x.ffill().bfill())
    df_pivot['ì¼ ìµœê³ ê¸°ì˜¨'] = df_pivot.groupby('ì˜ˆë³´ì¼ì')['ì¼ ìµœê³ ê¸°ì˜¨'].transform(lambda x: x.ffill().bfill())

    ordered_columns = [
        'ì˜ˆë³´ì¼ì', 'ì˜ˆë³´ì‹œê°', '1ì‹œê°„ ê¸°ì˜¨', 'í•˜ëŠ˜ìƒíƒœ', 'ìŠµë„', 'ê°•ìˆ˜í™•ë¥ ', 
        'ê°•ìˆ˜í˜•íƒœ', '1ì‹œê°„ ê°•ìˆ˜ëŸ‰', 'í’ì†', 'ì¼ ìµœì €ê¸°ì˜¨', 'ì¼ ìµœê³ ê¸°ì˜¨'
    ]
    final_columns = [col for col in ordered_columns if col in df_pivot.columns]
    
    return df_pivot[final_columns]

# ====================================================================
# 3. ë©”ì¸ ì‹¤í–‰ Ø¨Ø®Ø´ (Main Execution Section) - ì´ì „ê³¼ ë™ì¼
# ====================================================================
if __name__ == "__main__":
    nx, ny = get_grid_coords(TARGET_LOCATION_CITY, TARGET_LOCATION_DISTRICT, EXCEL_FILE_PATH)
    if nx and ny:
        print(f"'{TARGET_LOCATION_CITY} {TARGET_LOCATION_DISTRICT}'ì˜ ê²©ì ì¢Œí‘œ: nx={nx}, ny={ny}")
        
        forecast_raw_data = get_weather_forecast(SERVICE_KEY, nx, ny)
        
        if forecast_raw_data:
            df3 = create_forecast_dataframe(forecast_raw_data)
            
            pd.set_option('display.max_rows', None)
            pd.set_option('display.max_columns', None)
            pd.set_option('display.width', 1000)
            
           
```
